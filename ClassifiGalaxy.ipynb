### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2 as cv
import glob
from pathlib import Path
import tensorflow as tf
```

### Reading Data

```python
file=Path("images_gz2/images")
file
```

```python
images_path=list(file.glob(r"*.jpg"))
img_arr=[]
for i,data in enumerate(images_path[0:300]):
    img_arr.append(cv.imread(str(data)))
    cv.cvtColor(img_arr[i],cv.COLOR_BGR2RGB)
```

### Data Visualization

```python
plt.title("First Normal Galaxy")
plt.imshow(img_arr[0])
plt.plot()
```

```python
#Zoomed Galaxy
IMG_ANOTHER_GALAXY=img_arr[60]
plt.title("Zoomed On Galaxy")
plt.imshow(IMG_ANOTHER_GALAXY[100:300,100:300])
```

```python
fig,axs=plt.subplots(nrows=4,ncols=4,figsize=(18,8))
fig.tight_layout(pad=.5)
resized_img=[]
for i,ax in enumerate(axs.flat):
    ax.imshow(img_arr[i][100:300,100:300])
    ax.set_xlabel(np.asarray(img_arr).shape[0])
    resized_img.append(img_arr[i][100:300,100:300])
plt.plot()
```

```python
fig,axs=plt.subplots(nrows=4,ncols=4,figsize=(18,8))
fig.tight_layout(pad=.5)
image=[]
for i,ax in enumerate(axs.flat):
    image.append(cv.cvtColor(resized_img[i],cv.COLOR_RGB2GRAY))
    ax.imshow(image[i])
```

```python
kernel=np.array([[0,-1,0]
                ,[-1,5,-1]
                ,[0,-1,0]])
ed=cv.filter2D(resized_img[0],ddepth=-1,kernel=kernel)
canny_no_filter=cv.Canny(resized_img[5],5,80)
canny_filter=cv.Canny(ed,5,80)
fig,(ax1,ax2)=plt.subplots(2,figsize=(16,8))
ax1.imshow(canny_no_filter)
ax2.imshow(canny_filter)
```
